```python
#Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python
#Load the Dataset
df = pd.read_csv('train.csv')

df.head()
```

```python
# Data Cleaning 
df_cleaned = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])

df_cleaned['Age'] = df_cleaned['Age'].fillna(df_cleaned['Age'].median())
df_cleaned['Embarked'] = df_cleaned['Embarked'].fillna(df_cleaned['Embarked'].mode()[0])
```

```python
# Histogram of Age
plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned['Age'], bins=30, kde=True, color='skyblue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('SCT_DS_Task02_Histogram_Age.png')
plt.show()
```

```python
# Histogram of Fare
plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned['Fare'], bins=30, kde=True, color='orange')
plt.title('Fare Distribution of Passengers')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('SCT_DS_Task02_Histogram_Fare.png')
plt.show()
```

```python
# age distribution by class and gender
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_cleaned, x='Pclass', y='Age', hue='Sex', palette='Set2')
plt.title("Age Distribution by Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.legend(title="Sex")
plt.tight_layout()
plt.savefig("SCT_DS_Task02_Boxplot_Age_Pclass_Sex.png")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_cleaned, x='Sex', y='Age')
plt.title("Age Distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.tight_layout()
plt.savefig("SCT_DS_Task02_Boxplot_Age_Sex.png")
plt.show()
```

```python
# Survival Count by Sex
plt.figure(figsize=(7, 5))
sns.countplot(data=df_cleaned, x='Sex', hue='Survived', palette='pastel')
plt.title('Survival Count by Sex')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.savefig('SCT_DS_Task02_Countplot_Survival_Sex.png')
plt.show()
```

```python
# Survival Count by Passenger Class
plt.figure(figsize=(7, 5))
sns.countplot(data=df_cleaned, x='Pclass', hue='Survived', palette='muted')
plt.title('Survival Count by Passenger Class')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.savefig('SCT_DS_Task02_Countplot_Survival_Pclass.png')
plt.show()
```
