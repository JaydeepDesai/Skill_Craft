# SkillCraft Internship – Task 01: Population Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Total Population (2012–2022) ===

df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_395971.csv", skiprows=4)
```

```python
# Filter for India
india_data = df[df["Country Name"] == "India"]

# Select population values from 2012 to 2022
years = [str(year) for year in range(2012, 2023)]
india_population = india_data[years].values.flatten()
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(years, india_population / 1e9, color='teal')  # Convert to billions

# Add labels and title
plt.title("India's Population Growth (2012–2022)", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population (Billions)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate values on bars
for bar, pop in zip(bars, india_population):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, f"{pop/1e9:.2f}",
             ha='center', va='bottom', fontsize=9)

# Save chart
plt.tight_layout()
plt.savefig("SCT_DS_01_IndiaPopulationBarChart.jpg")
plt.show()
```

```python
# Simulated 2022 data for India (in millions)
age_groups = [
    "0-4", "5-9", "10-14", "15-19", "20-24", "25-29", "30-34",
    "35-39", "40-44", "45-49", "50-54", "55-59", "60-64",
    "65-69", "70-74", "75-79", "80+"
]
male_population = [65, 63, 60, 58, 56, 54, 52, 49, 47, 44, 42, 39, 36, 30, 25, 18, 12]
female_population = [61, 60, 57, 55, 53, 51, 49, 47, 45, 42, 40, 37, 34, 28, 23, 17, 11]

# Calculate total population per age group
total_population = [m + f for m, f in zip(male_population, female_population)]
```

```python
# Plot histogram using age group bars
plt.figure(figsize=(10, 6))
plt.bar(age_groups, total_population, color='purple')
plt.title("India's Population Histogram by Age Group (2022)", fontsize=14, fontweight='bold')
plt.xlabel("Age Group")
plt.ylabel("Population (Millions)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save and display
plt.tight_layout()
plt.savefig("SCT_DS_01_Histogram_AgeGroup2022.png")
plt.show()
```
